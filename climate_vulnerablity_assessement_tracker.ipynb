# Climate Change Vulnerability Assessment Tracker
Effects of climate change are already being seen everywhere -- unpredictable weathers, heatwaves, famine, biodiversity loss, and so on. Every city experiences a different set of adversities and affects the ecosystem and people differently. Climate vulnerability assessment takes into consideration the different built environments within a city or region that are more susceptible to climate related hazards such as floods, drought, heat waves, deteriorating air quality, impact of climate change, assesses the capacity of the entity to cope with and adapt to these impacts and identify areas of heightened vulnerabilities. This assessment enables policymakers to make informed decisions, resource allocation, build resilient strategies and reduce the risks associated with climate change. Rothman and Robinson (1997), have described the evolution of integrated vulnerability assessments through the following attributes, from linear to more complex chain of analyses, strictly quantitative to a more mixed analysis, from being science-driven to policy-driven assessments and from assessments that are dictated to the users to those that involve the users in the process.

One of the participants of the hackathon is interested in understanding how different countries do their vulnerability assessement. In this Notebook, we are using the #semanticClimate toolkit to mine the literature for various assesments reported in open scientific literature and create an interactive dashboard.

## Interactive Dashboard

```python
from pycountry_convert import country_alpha2_to_continent_code, country_name_to_country_alpha2
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
df_country_in_til_2 = pd.read_csv('country_til_human_annotated.csv')
```

```python
from geopy.geocoders import Nominatim
import folium
geolocator = Nominatim(user_agent="snh")
lat = []
long = []
for index, row in df_country_in_til_2.iterrows():
    ct = geolocator.geocode(row["country_names"], timeout=10)
    if ct is None:
        lat.append(None)
        long.append(None)
    else:
        latitude = (ct.latitude)
        longitude =(ct.longitude)
        lat.append(latitude)
        long.append(longitude)

df_country_in_til_2["lat"] = lat
df_country_in_til_2["long"]= long
```

```python
df = df_country_in_til_2.dropna()
```

```python
# import the library
import folium

# Make an empty map
m = folium.Map(location=[20,0], tiles="cartodb positron", zoom_start=2)

for i in range(0,len(df)):
    folium.Marker(location=[df.iloc[i]['lat'], df.iloc[i]['long']],popup=(df.iloc[i]['sentence'], df.iloc[i]['Floods/Other Disaster'],df.iloc[i]['file_path'].split('\\')[6])).add_to(m)

# Show the map again
m
```
