```python
from pycountry_convert import country_alpha2_to_continent_code, country_name_to_country_alpha2
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
df_country_in_til_2 = pd.read_csv('corpus/climate_vulnerability_assessment_floods_2023_2022/country_til.csv')
```

```python
from geopy.geocoders import Nominatim
import folium
geolocator = Nominatim(user_agent="snh")
lat = []
long = []
for index, row in df_country_in_til_2.iterrows():
    ct = geolocator.geocode(row["country_names"], timeout=10)
    if ct is None:
        lat.append(None)
        long.append(None)
    else:
        latitude = (ct.latitude)
        longitude =(ct.longitude)
        lat.append(latitude)
        long.append(longitude)

df_country_in_til_2["lat"] = lat
df_country_in_til_2["long"]= long
```

```python
df = df_country_in_til_2.dropna()
```

```python
# import the library
import folium

# Make an empty map
m = folium.Map(location=[20,0], tiles="cartodb positron", zoom_start=2)

for i in range(0,len(df)):
   folium.Marker(location=[df.iloc[i]['lat'], df.iloc[i]['long']],popup=df.iloc[i]['sentence'],).add_to(m)

# Show the map again
m
```
